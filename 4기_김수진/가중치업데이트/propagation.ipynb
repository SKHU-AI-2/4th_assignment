### 1. 순전파

```python
import numpy as np

def sigma(z):
    return 1 / (1 + np.exp(-z))

def diff_sigma(sigma_z):
    return sigma_z*(1 - sigma_z)

def mse(y, y_pred):
    return 1/2 * (y - y_pred)**2

x1, x2 = 0.1, 0.2
w1, w2, w3, w4 = 0.1, 0.2, 0.3, 0.15
w5, w6 = 0.25, 0.35
rl, y = 0.5, 0.3

```

```python
# 순전파
z1 = (x1*w1)+(x2*w2)
z2 = (x1*w3)+(x2*w4)

h1 = sigma(z1)
h2 = sigma(z2)

z3 = (h1*w5)+(h2*w6)
y_pred = sigma(z3)
```

```python
# 오차
loss = mse(y, y_pred)
```

```python
# 역전파
back1 = (y_pred - y)
back2 = diff_sigma(y_pred)
back3 = w5
back4 = diff_sigma(h1)
back5 = x1
res = back1*back2*back3*back4*back5

new_w1 = w1 - rl * res
```
